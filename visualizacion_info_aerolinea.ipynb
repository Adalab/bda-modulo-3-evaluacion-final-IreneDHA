```python
sns.countplot(x = 'Marital Status',
              hue='Gender', 
              data = df_aerolinea, 
              color = "turquoise");
```

```python
sns.countplot(x = "Loyalty Card", 
              data = df_aerolinea, 
              color = "brown");
```

```python
sns.barplot(x = "Education", 
            y = "Salary", 
            data = df_aerolinea,
            palette = "magma");
```

```python
#gracias a esta gráfica podemos entender el patrón de los nulos en 'Salary'
```

```python
sns.scatterplot(x = "Month", 
                y = "Total Flights", 
                data = df_aerolinea, 
                size = 5, 
                color = "green");

plt.title("Distribución de vuelos por mes");
plt.xlabel("Meses")
plt.ylabel("Número total de vuelos reservados");
```

```python
#el mes en el que más vuelos se reservan es julio. La temporada alta empieza en junio, termina en agosto y vuelve un poco en diciembre
```

```python
sns.scatterplot(x = "Distance", 
                y = "Points Accumulated", 
                data = df_aerolinea, 
                size = 5, 
                color = "blue")

plt.title("Relación entre la distancia y los puntos acumulados")
plt.xlabel("Distancia recorrida")
plt.ylabel("Puntos acumulados")
plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False);
```

```python
#sí, en general, cuanta más distancia, más puntos acumulas
```

```python
sns.histplot(x = "Province", 
             data = df_aerolinea, 
             color = "violet")

plt.title("Distribución de clientes por provincia")
plt.xlabel("Provincia")
plt.ylabel("Número de clientes")
plt.xticks(rotation=90)
plt.ylim(0, 150000);
```
