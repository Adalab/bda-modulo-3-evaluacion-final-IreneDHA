```python
#dataframes
import pandas as pd
pd.set_option('display.max_columns', None)
#operaciones matemáticas
import numpy as np
```

```python
info_vuelos = pd.read_csv("files/customer_flight_activity.csv")
info_loyalty = pd.read_csv("files/customer_loyalty_history.csv")
```

```python
info_vuelos.rename(columns=lambda x: x.title(), inplace=True)
#normalizamos todos los nombres
```

```python
info_vuelos.columns
```

```python
# Left merge para conservar todos los vuelos y agregar info de clientes acorde al loyalty number
df_aerolinea = pd.merge(info_vuelos, info_loyalty,on='Loyalty Number', how='left')
```

```python
#points redemmed - float o Accumulated - int #si bien sería bueno que fuesen el mismo tipo de dato, depende del sistema de la aerolinea (quizás solo permite gastar enteros, aunque puedas acumular decimales)
```

```python
df_aerolinea['Distance'] = df_aerolinea['Distance'].astype(float)
df_aerolinea['Dollar Cost Points Redeemed'] = df_aerolinea['Dollar Cost Points Redeemed'].astype(float)
#estas dos columnas, por la información que dan, deberían poder tener decimales
```

```python
df_aerolinea['Cancellation Year'] = df_aerolinea['Cancellation Year'].astype('Int64')
df_aerolinea['Cancellation Month'] = df_aerolinea['Cancellation Month'].astype('Int64')
```
